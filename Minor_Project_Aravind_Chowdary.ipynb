# 1.Importing the Dataset

```python
import numpy as np                                        #importing the pacakges required
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("datasets_4458_8204_winequality-red.csv")  #importing the dataset
df   #Pritning the dataset
```

# 2. Data Quality Check

```python
df=df[(df != 0).all(1)]  # droping all the rows that have atleast one null value and storing the remaining
df                       # Pritning the dataset after droping the rows with null values
```

# 3.Exploratory Data Analysis

## a. Univariate Anaylsis

### Histogarms

```python
col_list=sorted(df)               # Storing all the coloumn labels in one list in alphabetical order
for x in range(0,12):             # Iterating the loop for all the coloumn labels
    plt.hist(df[col_list[x]])
    plt.title(col_list[x])        # Pritning the title of the plot
    plt.show()                    # Plotting the Histogram
    plt.clf()               
```

### Box Plots

```python
for x in range(0,12):               # Iterating the loop for all the coloumn labels
    plt.boxplot(df[col_list[x]])
    plt.title(col_list[x])          # Pritning the title of the plot
    plt.show()                      # Plotting the Box plot
    plt.clf()
```

## b. Bivariate Analysis

### Scatter Plots

```python
for x in range(0,12):                # Iterating the loop for all the coloumn labels
    a=df[col_list[x]]
    y=df["quality"]
    plt.scatter(a,y)
    plt.xlabel(col_list[x])
    plt.ylabel("quality")
    plt.show()                        # Plotting the Scatter plot
    plt.clf()
```

# 4. MODELLING

## a. Train Test Split

```python
from sklearn.model_selection import train_test_split          #importing the pacakges required
X=df.drop(["quality"],axis=1).values
Y=df["quality"].values
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=40,stratify=Y)
# The test_size and random_state though alloted randomly, but are checked several times 
# whether they give a good test and train scores or not.
```

## b. KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier          #importing the pacakges required
n_neighbors=[]
for i in range(2,50):
    n_neighbors.append(i)
score = []
for a in n_neighbors:
    knn=KNeighborsClassifier(n_neighbors=a)
    knn.fit(X_train,Y_train)
    score.append(knn.score(X_test,Y_test))
m = score.index(max(score))
n = n_neighbors[m]
n
```

```python
knn=KNeighborsClassifier(n_neighbors=n)
knn.fit(X_train,Y_train)
knn.score(X_train,Y_train)                                  # Train score
```

```python
knn.score(X_test,Y_test)                                    # Test score
```

## d. Prediction on test set in KNN Classification

```python
Y_pred=knn.predict(X_test)
Y_pred                                                      # Y prediction values for X_test
```

## e. Accuracy in KNN Classification

```python
from sklearn.metrics import confusion_matrix                      #importing the pacakges required
from sklearn.metrics import classification_report
confusion_matrix(Y_test,Y_pred)                                   #Printing the Confusion matrix
```

```python
print(classification_report(Y_test,Y_pred))                       #Printing the Classification report  
```

## c. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression          #importing the pacakges required
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
logreg=LogisticRegression()
logreg.fit(X_train,Y_train)
logreg.score(X_train,Y_train)                                # Train score
```

```python
logreg.score(X_test,Y_test)                                  # Test score
```

## d. Prediction on test set in Logistic Regression

```python
Y_pred=logreg.predict(X_test)
Y_pred                                                       # Y prediction values for X_test
```

## e. Accuracy in Logistic Regression

```python
confusion_matrix(Y_test,Y_pred)                            #Printing the Confusion matrix
```

```python
print(classification_report(Y_test,Y_pred))               #Printing the Classification report 
```
